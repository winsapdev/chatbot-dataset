```python
lines = open('eng-ind.txt', encoding='utf-8').read().split('\n')
```

```python
print(len(lines))
```

```python
print(len(lines.read().split('/n')))
```

```python
eng_sentences = []
fra_sentences = []
eng_chars = set()
fra_chars = set()
nb_samples = 6752

# Process english and french sentences
for line in range(nb_samples):
    
    eng_line = str(lines[line]).split('\t')[0]
    
    # Append '\t' for start of the sentence and '\n' to signify end of the sentence
    fra_line = '\t' + str(lines[line]).split('\t')[1] + '\n'
    eng_sentences.append(eng_line)
    fra_sentences.append(fra_line)
    
    for ch in eng_line:
        if (ch not in eng_chars):
            eng_chars.add(ch)
            
    for ch in fra_line:
        if (ch not in fra_chars):
            fra_chars.add(ch)
            
# fra_sentences = fra_sentences.replace('\t', '').replace('\n', '')
```

```python
for idx in range(len(fra_sentences)):
    print(eng_sentences[idx])
    print(fra_sentences[idx].replace('\t', '').replace('\n', ''))
    print()
```
